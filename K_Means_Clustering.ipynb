```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()   #for plot styling
import numpy as np

# generating a two dimensional data set containing 5 distinct blobs
from sklearn.datasets import make_blobs
X,y_true=make_blobs(n_samples=400 ,centers=5 , cluster_std=0.6 , random_state=0)
plt.scatter(X[:,0],X[:,1])
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=5)
kmeans.fit(X)
y_kmeans=kmeans.predict(X)
```

```python
#visualization
plt.scatter(X[:,0],X[:,1],c=y_kmeans)
centers=kmeans.cluster_centers_
plt.scatter(centers[:,0],centers[:,1],c='blue',s=100)
```

```python
from sklearn.datasets import make_moons
X,y=make_moons(200,noise=.05,random_state=0)
plt.scatter(X[:,0],X[:,1])
```

```python
kmeans=KMeans(2,random_state=0).fit(X)
labels=kmeans.predict(X)
plt.scatter(X[:,0],X[:,1],c=labels,cmap='cool')
```
