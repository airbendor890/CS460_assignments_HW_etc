```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs

#Generate sample data
centers=[[1,1],[1,-1],[-1,1],[-1,-1]]
X,labels_true =make_blobs(n_samples=800,centers=centers,cluster_std=0.35,random_state=0)

plt.scatter(X[:,0],X[:,1],c=labels_true)
plt.title('sample clusters with true labels')
```

```python
#Run DBSCAN
#n=10,epsilon=0.3,euclidean
dbscan=DBSCAN(eps=0.3,min_samples=10).fit(X)

labels=dbscan.labels_
n_clusters_=len(set(labels))-(1 if -1 in labels else 0)
n_noise_=list(labels).count(-1)

print('Estimated number of clusters: %d' % n_clusters_)
print('Estimated number of noise points: %d' % n_noise_)

```

```python
plt.scatter(X[:,0],X[:,1],c=labels,cmap='cool')
plt.title('Estimated number of clusters: %d' % n_clusters_)
```
